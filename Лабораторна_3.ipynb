Для прикладу можна взяти цей код, який генерує список з 50 випадкових чисел в діапазоні від 1 до 100, обчислює їх суму та середнє значення, і виводить їх на екран.

```python
import random

# Генеруємо список випадкових чисел
random_numbers = [random.randint(1, 100) for _ in range(50)]

# Знаходимо суму чисел
total = sum(random_numbers)

# Обчислюємо середнє значення
average = total / len(random_numbers)

# Виводимо випадкові числа
print("Випадкові числа:")
for num in random_numbers:
    print(num)

# Виводимо суму та середнє значення
print(f"Сума чисел: {total}")
print(f"Середнє значення: {average}")
```

## **Робота з файлами**

***Читання файлів:***

Python надає простий спосіб читати дані з файлів та обробляти їх.

***Запис до файлів:***

Ви можете легко створювати та записувати дані до файлів з використанням Python.

***Оперування файловою системою:***

Python надає потужний набір інструментів для роботи з файловою системою, включаючи створення, переміщення, перейменування тощо.

Ось простий приклад коду для роботи з файлами. У цьому прикладі ми створимо файл, запишемо в нього деякий текст, прочитаємо цей текст та додамо до нього інший текст:



```python
# Створимо файл та запишемо в нього текст
with open("sample.txt", "w") as file:
    file.write("Це приклад тексту, який ми записали у файл.\n")
    file.write("Ще один рядок для нашого прикладу.\n")

# Прочитаємо вміст файлу
with open("sample.txt", "r") as file:
    file_contents = file.read()
    print("Вміст файлу:")
    print(file_contents)

# Допишемо ще рядок до файлу
with open("sample.txt", "a") as file:
    file.write("Це ще один рядок, який додали до файлу.")

# Знову прочитаємо вміст файлу, щоб переконатися, що рядок був доданий
with open("sample.txt", "r") as file:
    updated_contents = file.read()
    print("\nОновлений вміст файлу:")
    print(updated_contents)
```

У цьому коді ми використовуємо open() для створення, запису та читання файлу. З параметрами "w" ми відкриваємо файл для запису, "r" - для читання, і "a" - для дописування до файлу. Завжди важливо закривати файл після роботи з ним, і для цього ми використовуємо контекстний менеджер with.
